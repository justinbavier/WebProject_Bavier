***Data Cleaning***

Salary Column
- Displays '\n' after listed salaries, need to remove this. 
- Sometimes displays the letters 'A' or 'C' before listed salaries, need to remove this as well

```python
# Clean Salary column data
NoDupsDataNT['Salary'].replace({'\n': '' }, regex=True, inplace=True)
NoDupsDataNT['Salary'].replace({'A': '' }, regex=True, inplace=True)
NoDupsDataNT['Salary'].replace({'C': '' }, regex=True, inplace=True)
NoDupsData2NT['Salary'].replace({'\n': '' }, regex=True, inplace=True)
NoDupsData2NT['Salary'].replace({'A': '' }, regex=True, inplace=True)
NoDupsData2NT['Salary'].replace({'C': '' }, regex=True, inplace=True)
```

###### 3. Merge dataframes from separate json files as appropriate
- Find the intersection of two (or more) sets
- Compare the intersection with the newer set to find 'New Products'  
    - When found, add the starting date
- Compare the intersection with the older set to find 'Closed Products'  
    - When found, add the closing date
    - Compare closing date with starting date to find days on market      

```python
# Create the inner merge of the two dataframes
# Create 
dataInnerMerge = pd.merge(NoDupsDataNT, NoDupsData2NT, \
                          on=['Employer', 'Job Title', 'Location', 'Salary', 'Tags'],\
                          how='inner')
```

```python
# Appear on first and second days
# Jobs not taken

dataInnerMerge
```

```python
dataInnerMerge['BothDays'] = 1
dataInnerMerge
```

```python
# Appear on second day but not first
# New jobs
NoDupsData2NT['DayTwo'] = 1
newJob = pd.merge(NoDupsData2NT, dataInnerMerge,  \
                          on=['Employer', 'Job Title', 'Location', 'Salary', 'Tags'],\
                          how='left')

newJob = newJob[newJob.BothDays != newJob.DayTwo]

newJob
```

```python
# Appear on first day, but not second
# Closed jobs
NoDupsDataNT['DayOne'] = 1
closeJob = pd.merge(NoDupsDataNT, dataInnerMerge, \
                          on=['Employer', 'Job Title', 'Location', 'Salary', 'Tags'],\
                          how='left')

closeJob = closeJob[closeJob.DayOne != closeJob.BothDays]

closeJob
```

###### 5. Trends from your data  
- Search for specific brands and offer counts for each
- Search for product types and offer counts for each
- For each of the above (and other) give counts for day over day or week over week

```python
# List the products representing specific brands of your choosing
javaData = data[data['Tags'].str.contains('java ')]
phpData = data[data['Tags'].str.contains('php ')]

# Create a bar chart of products by brand (for the brands chosen)
javaCount = len(javaData)
phpCount = len(phpData)

objects = ('Java', '', 'PHP', '')
colors = ('green')

y_pos = np.arange(len(objects))
performance = [javaCount, 0, phpData, 0]
fig = plt.figure(figsize=(15,7))
plt.bar(y_pos, performance, align='center',alpha=0.5, color=colors)
plt.xticks(y_pos, objects)
plt.ylabel('Counts')
plt.title('Brands')
 
plt.show()

# List the products of a specific type


# Create a bar chart of products by specific type
# for each date provided
```
