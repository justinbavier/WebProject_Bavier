###### 3. Merge dataframes from separate json files as appropriate
- Find the intersection of two (or more) sets
- Compare the intersection with the newer set to find 'New Products'  
    - When found, add the starting date
- Compare the intersection with the older set to find 'Closed Products'  
    - When found, add the closing date
    - Compare closing date with starting date to find days on market      

```python
# Create the inner merge of the two dataframes
# Create 
dataInnerMerge = pd.merge(NoDupsDataNT, NoDupsData2NT, \
                          on=['Employer', 'Job Title', 'Location', 'Salary', 'Tags'],\
                          how='inner')
```

```python
# Appear on first and second days
# Jobs not taken

dataInnerMerge
```

```python
dataInnerMerge['BothDays'] = 1
dataInnerMerge
```

```python
# Appear on second day but not first
# New jobs
NoDupsData2NT['DayTwo'] = 1
newJob = pd.merge(NoDupsData2NT, dataInnerMerge,  \
                          on=['Employer', 'Job Title', 'Location', 'Salary', 'Tags'],\
                          how='left')

newJob = newJob[newJob.BothDays != newJob.DayTwo]

newJob
```

```python
# Appear on first day, but not second
# Closed jobs
NoDupsDataNT['DayOne'] = 1
closeJob = pd.merge(NoDupsDataNT, dataInnerMerge, \
                          on=['Employer', 'Job Title', 'Location', 'Salary', 'Tags'],\
                          how='left')

closeJob = closeJob[closeJob.DayOne != closeJob.BothDays]

closeJob
```

###### 4. Create visualizations
- Histogram of Prices
- Histogram of Days on Market
- Scatter Diagram of Prices vs Days on Market
- Pie Chart of New, Like New, Used

```python
# Jobs with specific languages
javaData = NoDupsDataNT[NoDupsData['Tags'].str.contains('java ')]
javascriptData = NoDupsDataNT[NoDupsData['Tags'].str.contains('javascript ')]
phpData = NoDupsDataNT[NoDupsData['Tags'].str.contains('php ')]
objCData = NoDupsDataNT[NoDupsData['Tags'].str.contains('objective-c ')]
swiftData = NoDupsDataNT[NoDupsData['Tags'].str.contains('swift ' )]
androidData = NoDupsDataNT[NoDupsData['Tags'].str.contains('android ' )]

# Count of each job
javaCount = len(javaData)
javascriptCount = len(javascriptData)
phpCount = len(phpData)
objCCount = len(objCData)
swiftCount = len(swiftData)
androidCount = len(androidData)

# Create a bar graph
objects = ('Java','', 'Javascript','', \
           'PHP','','Objective-C','','Swift','','Android')
colors = ("#4bc6fe",'red',"#41b8a9",'red', "#2e1366",'red', "#461c99",'red', "#5d26cc",'red', "#7530ff")
y_pos = np.arange(len(objects))
performance = [javaCount,0,javascriptCount,\
               0,phpCount,0,objCCount,0,swiftCount,0,androidCount]
fig = plt.figure(figsize=(15,7))
plt.bar(y_pos, performance, align='center',alpha=0.5, color=colors)
plt.xticks(y_pos, objects)
plt.ylabel('Counts')
plt.title('Language')

plt.show()
```

```python
# Create a Pie Chart of New, Like New, and Used products
colors = ["#4bc6fe", "#41b8a9", "#2e1366", "#461c99", "#5d26cc", "#7530ff"]
# Create a pie chart
plt.pie(
    [NoDupsDataNT.Tags[NoDupsData['Tags'].str.contains('java ')].count(),\
     NoDupsDataNT.Tags[NoDupsData['Tags'].str.contains('javascript ')].count(), \
     NoDupsDataNT.Tags[NoDupsData['Tags'].str.contains('php ')].count(), \
    NoDupsDataNT.Tags[NoDupsData['Tags'].str.contains('objective-c ')].count(), \
    NoDupsDataNT.Tags[NoDupsData['Tags'].str.contains('swift ')].count(), \
    NoDupsDataNT.Tags[NoDupsData['Tags'].str.contains('android ')].count()], \
    labels= ['Java','Javascript', 'PHP', 'Objective-C', 'Swift', 'Android'],
    shadow=True,
    # with colors
    colors=colors,
    # with one slide exploded out
    explode=(0.05, 0.05, 0.05, 0.05, 0.05, 0.05),
    # with the start angle at 90%
    startangle=180,
    # with the percent listed as a fraction
    autopct='%1.1f%%',
    )

# View the plot drop above
plt.axis('equal')

# View the plot
plt.tight_layout()
plt.show()
```
